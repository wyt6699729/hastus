```python
my_vol_path = '/Volumes/sdp_advance/multiplex_1_bronze'
```

```python
spark.sql(f"LIST '/Volumes/sdp_advance/multiplex_1_bronze/business_events' ")
```

```python
spark.sql(f"""
SELECT
  CAST(key AS STRING) AS key,
  CAST(value AS STRING) AS value_str,
  topic,
  partition,
  offset
FROM read_files('{my_vol_path}/business_events')
LIMIT 10
""")
```

```python
spark.sql(f"""
SELECT
  CAST(key AS STRING) AS key,
  CAST(value AS STRING) AS value_str,
  topic,
  partition,
  offset
FROM read_files('{my_vol_path}/business_events')
LIMIT 10
""")
```

```python
spark.sql(f"""
WITH raw_events AS (
  SELECT
    CAST(value AS STRING) AS value_str
  FROM read_files('{my_vol_path}/business_events')
),
parsed_events AS (
  SELECT
    value_str,
    parse_json(value_str) AS event_data_variant
  FROM raw_events
)
SELECT
  value_str,
  event_data_variant,
  CAST(event_data_variant:event_id AS STRING) AS extracted_event_id,
  event_data_variant:timestamp::TIMESTAMP AS extracted_timestamp
FROM parsed_events
LIMIT 10
""")
```

```python
# a. Total row count
df_count = spark.sql(f"""
    SELECT COUNT(*) AS total_rows
    FROM  read_files('{my_vol_path}/business_events')
""")
display(df_count)

# b. Data source count by topic
df_data_source_count = spark.sql(f"""
    SELECT 
        topic,
        COUNT(*) AS total_rows
    FROM read_files('{my_vol_path}/business_events')
    GROUP BY topic
    ORDER BY topic
""")
display(df_data_source_count) 
```
